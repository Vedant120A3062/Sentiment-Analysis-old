```python
from textblob import TextBlob
from googletrans import Translator
```

```python
# Function to translate text
def translate_text(text, target_language='en'):
    translator = Translator()
    translated = translator.translate(text, dest=target_language)
    return translated.text
```

```python
# Function to perform sentiment analysis
def analyze_sentiment(text):
    blob = TextBlob(text)
    sentiment = blob.sentiment

    if sentiment.polarity > 0:
        return "Positive"
    elif sentiment.polarity < 0:
        return "Negative"
    else:
        return "Neutral"
```

```python
# Input text to be analyzed
input_text = input("Enter text to analyze sentiment: ")
```

```python
# Translate the text to English (or your desired target language)
translated_text = translate_text(input_text, target_language='en')
print("Translated Text:", translated_text)
```

```python
# Perform sentiment analysis on the translated text
sentiment = analyze_sentiment(translated_text)
print("Sentiment:", sentiment)
```
